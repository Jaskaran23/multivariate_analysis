## Correlation plot
#### below plot have been plotted to see the correlation between price and other data variables in the data. As we can see that all data variables have correlation between 0.6 and 1.0 given the lighter color except some features such as square feet of lot. We can utilize this inforamtion while training our model and can include and exclude these variables to see the effect on the accuracy. This has been done due to the reason that multivariate regression's result sometimes deteriorate due to inclusion of many features, therefore one has to identify the features which will have the most impact

```python
corr = df1.corr()
plt.pyplot.subplots(figsize=(10,8))
sns.heatmap(corr, square=True)
```
